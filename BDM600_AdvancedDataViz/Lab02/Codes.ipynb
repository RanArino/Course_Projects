```python
import pandas as pd
```

```python
# load data
origin = pd.read_csv('Lab02/data.csv')
data = origin[['Platform', 'Genre', "NA_Sales"]] 
data.head()
```

```python
# focus on the top 10 Platform
data = data[data['Platform'].isin(list(data['Platform'].value_counts()[:10].keys()))].reset_index(drop=True)
data.head()
```

```python
#data.set_index('Platform').groupby('Genre').mean()
bertifier_df = data.pivot_table(index='Genre', columns='Platform', values='NA_Sales', aggfunc='mean')
bertifier_df
```

```python
bertifier_df.to_csv('bertifier_df.csv')
```
