```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def custom_nonlinear_model(x, a,b,c):
    return a * np.sin(b * x) + c * np.exp(-0.1 * x)
```

```python
np.random.seed(0)
x_data = np.linspace(-10, 10, 100)
y_data = custom_nonlinear_model(x_data, 2.5, 1.3, 0.5) + 0.2 * np.random.randn(100)

#print(np.mean(np.random.randn(100)))
#print(np.std(np.random.randn(100))**2)
```

```python
# mean and varience
print("mean: ", np.mean(x_data))
print("varience: ", np.std(x_data)**2)
```

```python
params, covarience = curve_fit(custom_nonlinear_model, x_data, y_data, p0=[2, 1, 1])
```

```python
a_opt, b_opt, c_opt = params
print(a_opt, b_opt, c_opt)
```

```python
y_fit = custom_nonlinear_model(x_data, a_opt, b_opt, c_opt)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(x_data, y_data, label='Data', color='blue', alpha=0.5)
plt.plot(x_data, y_fit, label='Fitted Curve', color='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Advanced Nonlinear Regression Example')
plt.legend()
plt.grid(True)
```
